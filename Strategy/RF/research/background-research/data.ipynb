## Data: Alpaca API
---
For single symbol only

```python
import requests
import pandas as pd

from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv('API_KEY')
secret_key = os.getenv('SECRET_KEY')
```

```python
url = "https://paper-api.alpaca.markets/v2/assets?attributes="

headers = {
    "accept": "application/json",
    "APCA-API-KEY-ID": api_key,
    "APCA-API-SECRET-KEY": secret_key,
}

response = requests.get(url, headers=headers)
```

```python
from alpaca.data.live import StockDataStream


wss_client = StockDataStream(api_key,secret_key)

# async handler
async def bars_data_handler(data):
    # quote data will arrive here
    print(data)

wss_client.subscribe_bars(bars_data_handler, ["AAPL","SPY"])

wss_client.run()
```

```python
data = response.json()  # Parse the JSON response
df = pd.DataFrame(data)  # Load the data into a pandas DataFrame
df.index = df['id']
df.drop('id', axis=1, inplace=True)
df.head()
```
